```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
amz_df = pd.read_csv("./Data/amazon_delivery.csv")
amz_df
```

# Removing Some Columns

```python
amz_df.drop(columns=['Agent_Age',"Category"],inplace=True)#droping irrelevent columns
```

```python
amz_df['Agent_Rating'] = amz_df['Agent_Rating'].fillna(amz_df['Agent_Rating'].mean())
amz_df
```

# Performing Ordinal Encoding

```python
print(amz_df['Traffic'].unique()) #to check  all the unique values in the column
print(amz_df['Weather'].unique())
amz_df['Traffic'] = amz_df['Traffic'].str.strip()#removing extra spaces so that we don't get problems during mapping
amz_df['Weather'] = amz_df['Weather'].str.strip()
ordinal_map = {'Low': 1, 'Medium': 2, 'High': 3, 'Jam':4, 'NaN': 3}
ord_map = {'Sunny' :1, 'Cloudy':1, 'Windy':1 ,'Stormy': 3, 'Sandstorms' :2, 'Fog':4, 'nan':3}

ordinal_encoded_df = amz_df.copy()
ordinal_encoded_df['Traffic_Ordinal'] = amz_df['Traffic'].map(ordinal_map)#.fillna(-1) ## # or another placeholder value
ordinal_encoded_df['Weather_Ordinal'] = amz_df['Weather'].map(ord_map)
#ordinal_encoded_df.drop(['Traffic','Order_ID','Vehicle','Area','Weather', 'Order_Time','Pickup_Time','Order_Date'],axis=1, inplace= True)#Removing some more irrelevant 
ordinal_encoded_df['Delivery_Time'] = ordinal_encoded_df['Delivery_Time'].fillna(ordinal_encoded_df['Delivery_Time'].mean())
ordinal_encoded_df
```

```python
ordinal_encoded_df.to_csv("routeCity.csv",index=False)
```
